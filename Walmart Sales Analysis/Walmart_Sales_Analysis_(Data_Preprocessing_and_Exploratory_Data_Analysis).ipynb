# Outlier Detection and Handling:

```python
# Detect and handle outliers using the IQR method

Q1 = df_train['Weekly_Sales'].quantile(0.25)
Q3 = df_train['Weekly_Sales'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outliers = df_train[(df_train['Weekly_Sales'] < lower_bound) | (df_train['Weekly_Sales'] > upper_bound)]
df_train_cleaned = df_train[~df_train.index.isin(outliers.index)]

df_train_cleaned
```
