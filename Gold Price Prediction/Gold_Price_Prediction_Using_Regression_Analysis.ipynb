# Splitting the Data into Training and Testing Set for Modeling:

```python
# Assuming df is your DataFrame and 'target_column' is the column you want as y
target_column_name = 'Adj Close'

# Separate y (target variable)
y = df[target_column_name]

# Separate X (features) by excluding the target column
x = df.drop(columns=[target_column_name])
x = x.drop(columns= ['Date','Open','High','Low','Close','Volume'])
```

```python
# Train Test Split

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size = 0.20, random_state =42)
```

```python
x_test
```

***Note:*** Notice the structure of the x_test variable. this is the structure we have to input for unseen data in the future after we select the final model.

# Feature Scaling:

```python


from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
sc_x_train = scaler.fit_transform(x_train)
sc_x_test= scaler.transform(x_test)


# Assuming y_train and y_test are Pandas Series
y_train = y_train.values.reshape(-1, 1)

sc_y_train = scaler.fit_transform(y_train)
```

```python
sc_x_test
```

# Principal Component Analysis:

```python
from sklearn.decomposition import PCA
pca = PCA()
pca.fit(sc_x_train)
cumsum = np.cumsum(pca.explained_variance_ratio_)
d  = np.argmax(cumsum >= 0.95 ) + 1

pca = PCA(n_components = d)
x_train_pca = pca.fit_transform(sc_x_train)
x_test_pca = pca.transform(sc_x_test)
```

# Linear Regression:

```python
# Model creation and prediction

from sklearn.linear_model import LinearRegression
lr_model = LinearRegression()
lr_model.fit(x_train_pca,y_train)
lr_y_pred = scaler.inverse_transform(lr_model.predict(x_test_pca))

# Model Evaluation
from sklearn.metrics import mean_squared_error,r2_score
mse = mean_squared_error(y_test, lr_y_pred)
rmse = np.sqrt(mse)
r_square = r2_score(y_test, lr_y_pred)
print('RMSE =', rmse)
print('R-Square = ', r_square)
```

```python
# K-Fold Cross validation

from sklearn.model_selection import cross_val_score
accuracies  = cross_val_score(estimator = lr_model, X = x_train_pca, y = y_train, cv =10)

print('Average Accuracy = ',accuracies.mean())
print('St. Dev. of Accuracy = ',accuracies.std())
```

#Support Vector Regression (Linear)

```python
from sklearn.svm import SVR
svr_lin = SVR(kernel='linear')
svr_lin.fit(x_train_pca,y_train)
svr_lin_y_pred = scaler.inverse_transform((svr_lin.predict(x_test_pca)).reshape(-1,1))

# Model Evaluation
from sklearn.metrics import mean_squared_error,r2_score
mse = mean_squared_error(y_test, svr_lin_y_pred)
rmse = np.sqrt(mse)
r_square = r2_score(y_test, svr_lin_y_pred)
print('RMSE =', rmse)
print('R-Square = ', r_square)
```

```python
# K-Fold Cross validation

from sklearn.model_selection import cross_val_score
accuracies  = cross_val_score(estimator =svr_lin , X = x_train_pca, y = y_train.ravel(), cv =10)

print('Average Accuracy = ',accuracies.mean())
print('St. Dev. of Accuracy = ',accuracies.std())
```

# Random Forest Regressor:

```python
from sklearn.ensemble import RandomForestRegressor
RFR = RandomForestRegressor()
RFR.fit(x_train_pca,y_train)
RFR_y_pred = scaler.inverse_transform((RFR.predict(x_test_pca)).reshape(-1,1))

# Model Evaluation
from sklearn.metrics import mean_squared_error,r2_score
mse = mean_squared_error(y_test, RFR_y_pred)
rmse = np.sqrt(mse)
r_square = r2_score(y_test, RFR_y_pred)
print('RMSE =', rmse)
print('R-Square = ', r_square)
```

```python
# K-Fold Cross validation

from sklearn.model_selection import cross_val_score
accuracies  = cross_val_score(estimator = RFR, X = x_train_pca, y = y_train, cv =10)

print('Average Accuracy = ',accuracies.mean())
print('St. Dev. of Accuracy = ',accuracies.std())
```

# XGBoost Regression:

```python
from xgboost import XGBRegressor
XGBR = XGBRegressor()
XGBR.fit(x_train_pca,y_train)
XGBR_y_pred = scaler.inverse_transform((XGBR.predict(x_test_pca)).reshape(-1,1))

# Model Evaluation
from sklearn.metrics import mean_squared_error,r2_score
mse = mean_squared_error(y_test, XGBR_y_pred)
rmse = np.sqrt(mse)
r_square = r2_score(y_test, XGBR_y_pred)
print('RMSE =', rmse)
print('R-Square = ', r_square)
```

```python
# K-Fold Cross validation

from sklearn.model_selection import cross_val_score
accuracies  = cross_val_score(estimator = XGBR, X = x_train_pca, y = y_train, cv =10)

print('Average Accuracy = ',accuracies.mean())
print('St. Dev. of Accuracy = ',accuracies.std())
```

So, we finally select the XGBoost Regressor Model to predict Gold Price as it gives the Highest Accuracy (i.e lowest Root Mean Squared Error score)

Now, let's try and see if we can predict Gold Price using Artificial Neural Networks.

# Artificial Neural Network (Regression):

```python
import tensorflow as tf

ann = tf.keras.models.Sequential()
ann.add(tf.keras.layers.Dense(units=40, activation='relu'))
ann.add(tf.keras.layers.Dense(units=40, activation='relu'))
ann.add(tf.keras.layers.Dense(units=1, activation='linear'))  # Using 'linear' activation for regression

ann.compile(optimizer='adam', loss='mean_squared_error')  # Using mean squared error for regression
ann.fit(x_train_pca, y_train, batch_size=32, epochs=10)
```

```python
loss_df = pd.DataFrame(ann.history.history)
loss_df.plot() # to know how many epochs is enough once loss plot becomes flat
```

```python
ann_y_pred = scaler.inverse_transform(ann.predict(x_test_pca))

# Model Evaluation
from sklearn.metrics import mean_squared_error,r2_score
mse = mean_squared_error(y_test, ann_y_pred)
rmse = np.sqrt(mse)
r_square = r2_score(y_test, ann_y_pred)
print('RMSE =', rmse)
print('R-Square = ', r_square)
```

***NOTE:*** Here, i've been facing many problems with inputting an unseen data containing only one row. The min-max or standard scaler keeps giving error because it needs more than one row of data to determine the min,max,mean,std values for feature scaling. so, i've done the scaling manually.  

```python
import pandas as pd

def predict_input(new_input, x_train):

    # Convert x_train to DataFrame if it's a NumPy array
    if isinstance(x_train, np.ndarray):
        x_train = pd.DataFrame(x_train, columns=list(new_input.keys()))

    # Calculate mean and standard deviation for each feature in x_train
    feature_means = x_train.mean(axis=0)
    feature_std_devs = x_train.std(axis=0)

    # Standard scaling for your one row of data
    for col in new_input.keys():
        if feature_std_devs[col] != 0:
            new_input[col] = (new_input[col] - feature_means[col]) / feature_std_devs[col]

    x_input_scaled = pd.DataFrame([new_input])[list(new_input.keys())]

    # Assuming you have already defined model_scaler, model_pca, and selected_model
    x_input_pca = model_pca.transform(x_input_scaled)

    # Make predictions using your trained model
    pred = selected_model.predict(x_input_pca)[0]

    return pred
```
