# Splitting the Data into Training and Testing Set for Modeling:

```python
# Assuming df is your DataFrame and 'target_column' is the column you want as y
target_column_name = 'Adj Close'

# Separate y (target variable)
y = df[target_column_name]

# Separate X (features) by excluding the target column
x = df.drop(columns=[target_column_name])
x = x.drop(columns= ['Date','Open','High','Low','Close','Volume'])
```

```python
# Train Test Split

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size = 0.20, random_state =42)
```

```python
x_test
```

***Note:*** Notice the structure of the x_test variable. this is the structure we have to input for unseen data in the future after we select the final model.

# Feature Scaling:

```python


from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
sc_x_train = scaler.fit_transform(x_train)
sc_x_test= scaler.transform(x_test)


# Assuming y_train and y_test are Pandas Series
#y_train = y_train.values.reshape(-1, 1)

#sc_y_train = scaler.fit_transform(y_train)
```

```python
x_train.shape # checking the shape inputted into scaler because it is causing problems for future unseen data
```

```python
sc_x_train.shape
```

```python
y_test  # we want the results of regression models in this format to find RMSE Values
```

# Principal Component Analysis:

```python
from sklearn.decomposition import PCA
pca = PCA()
pca.fit(sc_x_train)
cumsum = np.cumsum(pca.explained_variance_ratio_)
d  = np.argmax(cumsum >= 0.95 ) + 1

pca = PCA(n_components = d)
x_train_pca = pca.fit_transform(sc_x_train)
x_test_pca = pca.transform(sc_x_test)
```

# Linear Regression:

***Note***: y_pred value will be in the format of y_train in the model. if you scale y_train , you will get a scaled version as output in y_pred. So, in this case you have to inverse_transform the y_pred value to get the results in the original units.

```python
# Model creation and prediction

from sklearn.linear_model import LinearRegression
lr_model = LinearRegression()
lr_model.fit(x_train_pca,y_train)
lr_y_pred = lr_model.predict(x_test_pca)

# Model Evaluation
from sklearn.metrics import mean_squared_error,r2_score
mse = mean_squared_error(y_test, lr_y_pred)
rmse = np.sqrt(mse)
r_square = r2_score(y_test, lr_y_pred)
print('RMSE =', rmse)
print('R-Square = ', r_square)
```

```python
lr_y_pred
```

```python
# K-Fold Cross validation
from sklearn.metrics import mean_squared_error, make_scorer
from sklearn.model_selection import cross_val_score

# Define a custom scoring function for RMSE
rmse_scorer = make_scorer(mean_squared_error, squared=False)
# Perform k-fold cross-validation with RMSE scoring
accuracies = cross_val_score(estimator=lr_model, X=x_train_pca, y=y_train, cv=10, scoring=rmse_scorer)

# Print the results
print('Average RMSE =', accuracies.mean())
print('Standard Deviation of RMSE =', accuracies.std())
```

#Support Vector Regression (Linear)

```python
from sklearn.svm import SVR
svr_lin = SVR(kernel='linear')
svr_lin.fit(x_train_pca,y_train)
svr_lin_y_pred = svr_lin.predict(x_test_pca)

# Model Evaluation
from sklearn.metrics import mean_squared_error,r2_score
mse = mean_squared_error(y_test, svr_lin_y_pred)
rmse = np.sqrt(mse)
r_square = r2_score(y_test, svr_lin_y_pred)
print('RMSE =', rmse)
print('R-Square = ', r_square)
```

```python
# K-Fold Cross validation
from sklearn.metrics import mean_squared_error, make_scorer
from sklearn.model_selection import cross_val_score

# Define a custom scoring function for RMSE
rmse_scorer = make_scorer(mean_squared_error, squared=False)
# Perform k-fold cross-validation with RMSE scoring
accuracies = cross_val_score(estimator=svr_lin, X=x_train_pca, y=y_train, cv=10, scoring=rmse_scorer)

# Print the results
print('Average RMSE =', accuracies.mean())
print('Standard Deviation of RMSE =', accuracies.std())
```

# Random Forest Regressor:

```python
from sklearn.ensemble import RandomForestRegressor
RFR = RandomForestRegressor()
RFR.fit(x_train_pca,y_train)

RFR_y_pred = RFR.predict(x_test_pca)
# Model Evaluation
from sklearn.metrics import mean_squared_error,r2_score
mse = mean_squared_error(y_test, RFR_y_pred)
rmse = np.sqrt(mse)
r_square = r2_score(y_test, RFR_y_pred)
print('RMSE =', rmse)
print('R-Square = ', r_square)
```

```python
# K-Fold Cross validation
from sklearn.metrics import mean_squared_error, make_scorer
from sklearn.model_selection import cross_val_score

# Define a custom scoring function for RMSE
rmse_scorer = make_scorer(mean_squared_error, squared=False)
# Perform k-fold cross-validation with RMSE scoring
accuracies = cross_val_score(estimator=RFR, X=x_train_pca, y=y_train, cv=10, scoring=rmse_scorer)

# Print the results
print('Average RMSE =', accuracies.mean())
print('Standard Deviation of RMSE =', accuracies.std())
```

# Artificial Neural Network (Regression):

```python
import tensorflow as tf

ann = tf.keras.models.Sequential()
ann.add(tf.keras.layers.Dense(units=40, activation='relu'))
ann.add(tf.keras.layers.Dense(units=40, activation='relu'))
ann.add(tf.keras.layers.Dense(units=1, activation='linear'))  # Using 'linear' activation for regression

ann.compile(optimizer='adam', loss='mean_squared_error')  # Using mean squared error for regression
ann.fit(x_train_pca, y_train, batch_size=32, epochs=5)
```

```python
loss_df = pd.DataFrame(ann.history.history)
loss_df.plot() # to know how many epochs is enough once loss plot becomes flat
```

```python

ann_y_pred = ann.predict(x_test_pca)
# Model Evaluation
from sklearn.metrics import mean_squared_error,r2_score
mse = mean_squared_error(y_test, ann_y_pred)
rmse = np.sqrt(mse)
r_square = r2_score(y_test, ann_y_pred)
print('RMSE =', rmse)
print('R-Square = ', r_square)
```
