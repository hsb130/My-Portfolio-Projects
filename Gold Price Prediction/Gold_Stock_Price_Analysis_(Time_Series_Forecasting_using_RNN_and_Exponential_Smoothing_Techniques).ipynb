# Daily Returns (%) Calculation:

We assume an investor started investing on Gold during the start date of this data set. We will do Normalization on the dataset to set the initial stock price at $1 ,to calculate how much return he is getting for each Dollar he invested.  

````python
# print out normalized (scaled) stock prices

def normalize(df):
  x = df.copy() # make a copy of my dataframe
  for i in x.columns[:]:
    x[i] = x[i]/(x[i][0]) #taking all the values in a column and dividing by the first value of that column
  return x
# now call the function
df_2 = normalize(df)
df_2
````

````python
# Plotting the 'Adj Close' over time
plt.figure(figsize=(20, 6))
plt.plot(df_2['Adj Close'])
plt.title('Daily Adjusted Close Prices Over Time')
plt.xlabel('Date')
plt.ylabel('Adj Close')
plt.show()
````

````python
# Plotting the 'Volume' over time
plt.figure(figsize=(20, 6))
plt.plot(df_2['Volume'])
plt.title('Daily Volume Over Time')
plt.xlabel('Date')
plt.ylabel('Adj Close')
plt.show()
````

````python
# Create a new column for daily returns
df_2['Daily Return'] = ((df_2['Adj Close'] - df_2['Adj Close'].shift(1)) / df_2['Adj Close'].shift(1)) * 100

# Set the first value of the 'Daily Return' column to 0
df_2['Daily Return'].iloc[0] = 0

# Print the DataFrame with the new 'Daily Return' column
print(df_2)
````

````python
# Plotting the 'Daily Returns' over time
plt.figure(figsize=(20, 6))
plt.plot(df_2['Daily Return'])
plt.title('Daily Return % Over Time')
plt.xlabel('Date')
plt.ylabel('Adj Close')
plt.show()
````

# Time Series Forecasting:

Screenshot (11).png

# Step-1: Check for Trend and Seasonality:

````python
from statsmodels.tsa.seasonal import seasonal_decompose


# Decompose the time series into trend, seasonal, and residual components
result = seasonal_decompose(df['Adj Close'], model='additive', period= 365)

# Plot the decomposed components
plt.figure(figsize=(20, 8))
result.plot()
plt.title('Seasonal Decomposition of Time Series')
plt.show()

# Check if there is a trend
if result.trend is not None:
    print("The time series has a trend.")
else:
    print("The time series does not have a trend.")

# Check if there is seasonality
if result.seasonal is not None:
    print("The time series has seasonality.")
else:
    print("The time series does not have seasonality.")
````

So, We will use Holt Winter's Additive Seasonality Model with Trend.

# Step-2: Forecasting using Holt Winters Seasonality Model with Trend:



````python
from statsmodels.tsa.holtwinters import ExponentialSmoothing


ts = df['Adj Close']

# Split the data into training and testing sets
# You can adjust the split point based on your specific requirements
train_size = int(len(ts) * 0.8)
train, test = ts[:train_size], ts[train_size:]

# Build the Holt-Winters Additive Seasonality model with trend
model = ExponentialSmoothing(train, trend='add', seasonal='add', seasonal_periods=365)
result = model.fit()

# Make predictions for the test set
predictions = result.forecast(len(test))

# Plot the original time series and the forecast
plt.figure(figsize=(20, 10))
plt.plot(train.index, train, label='Training Data')
plt.plot(test.index, test, label='Testing Data')
plt.plot(test.index, predictions, label='Forecast')
plt.title('Holt-Winters Additive Seasonality Forecast')
plt.xlabel('Date')
plt.ylabel('Adjusted Closing Price')
plt.legend()
plt.show()
````

````python
# Model Evaluation:

from sklearn.metrics import mean_squared_error
# Calculate the Mean Squared Error (MSE)
mse = mean_squared_error(test, predictions)

# Calculate the Root Mean Squared Error (RMSE)
rmse = np.sqrt(mse)


print(f"Root Mean Squared Error (RMSE): {rmse}")

````

# Step-3: Forecasting using Recurrent Neural Networks:

````python
train.shape
````

````python
test.shape
````

````python
X_train = []
y_train = []

# let's take the first 1374 (80 %) datapoints as training data

for i in range(30, len(train)):
    X_train.append(train[i-30:i]) # i want to forecast next day gold price (based on 30 days (lag) data of the past)
    y_train.append(train[i])


X_train, y_train = np.array(X_train), np.array(y_train)
````

````python
# Reshaping Data for inputing into Keras

X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], 1))
````

***NOTE:***
Here, x and y are 2D array(observation and features) into a 3D array where the 3rd dimension will include the timestamp.

why do we do this?
Keras requires 3D tensor with shape (batch_size, timestamps, input_dim) as input. here, our input_dim is =1 (stock price at time 't')

````python
# now, lets take remaining 20% of data as testing set

X_test = []
for i in range(30,len(test)):
    X_test.append(test[i-30:i])

X_test = np.array(X_test)
X_test = np.reshape(X_test, (X_test.shape[0], X_test.shape[1], 1))
````

````python
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from keras.layers import Dropout

regressor = Sequential()

# Input Layer
regressor.add(LSTM(units = 50, return_sequences = True, input_shape = (X_train.shape[1], 1)))
regressor.add(Dropout(0.2))

# 1st Hidden Layer
regressor.add(LSTM(units = 50, return_sequences = True))
regressor.add(Dropout(0.2))

# 2nd Hidden Layer
regressor.add(LSTM(units = 50, return_sequences = True))
regressor.add(Dropout(0.2))

# 3rd Hidden Layer
regressor.add(LSTM(units = 50))
regressor.add(Dropout(0.2))

# Output Layer
regressor.add(Dense(units = 1))

regressor.compile(optimizer = 'adam', loss = 'mean_squared_error')

regressor.fit(X_train, y_train, epochs = 500, batch_size = 32)
````

````python
pred = regressor.predict(X_test)
````

````python

# Reshape predictions to 1D array
y_pred_1d = pred.flatten()

# Plot the training, testing, and forecasted data
plt.figure(figsize=(20, 10))
plt.plot(train.index, train, label='Training Data')
plt.plot(test.index, test, label='Testing Data')
plt.plot(test.index[:-30], y_pred_1d, label='Forecast')  # Adjust indexing to match lengths
plt.title('Forecast using RNN Model')
plt.xlabel('Date')
plt.ylabel('Value')
plt.legend()
plt.show()
````

````python
# Reshape predictions and true values to 1D arrays
y_true_1d = test.values[:len(y_pred_1d)].flatten()

# Calculate Root Mean Squared Error (RMSE)
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_true_1d, y_pred_1d)
rmse = np.sqrt(mse)

print(f"Root Mean Squared Error (RMSE): {rmse}")
````

Now, we can finally decide which model to use. RNN is the winner here as it's RMSE value is very low compared to the Holt-Winter's Seasonal Model with trend.
