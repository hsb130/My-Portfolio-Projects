# Necessary Libraries and Datasets import:

```python
import numpy as np
import pandas as pd

from google.colab import drive
drive.mount('/content/drive')
```

```python
content_df = pd.read_csv('/content/drive/MyDrive/Data Science Projects/Accenture North America Data Analysis Project/Datasets/Content.csv')

reactions_df = pd.read_csv('/content/drive/MyDrive/Data Science Projects/Accenture North America Data Analysis Project/Datasets/Reactions.csv')

reaction_type_df = pd.read_csv('/content/drive/MyDrive/Data Science Projects/Accenture North America Data Analysis Project/Datasets/ReactionTypes.csv')
```

# Dropping unnecessary columns:

```python
content_df.drop(columns=['Unnamed: 0','User ID','URL'], inplace = True)
```

```python
content_df
```

```python
reactions_df.drop(columns = ['Unnamed: 0','User ID'], inplace = True)
reactions_df
```

```python
reaction_type_df.drop(columns=['Unnamed: 0'], inplace=True)
reaction_type_df
```

# Changing Column Names for ease of merging:

```python
content_df.rename(columns={'Type': 'Content Type'}, inplace=True)
content_df
```

```python
reactions_df.rename(columns={'Type': 'Reaction Type'}, inplace=True)
reactions_df
```

```python
reaction_type_df.rename(columns={'Type': 'Reaction Type'}, inplace=True)
reaction_type_df
```

# Merging the 3 Datasets into one Dataframe:

```python
# Merge dataframes (using inner join)
merged_df = pd.merge(content_df, reactions_df, on='Content ID')
merged_df = pd.merge(merged_df, reaction_type_df, on='Reaction Type')

merged_df
```

# Missing Values Handling:

```python
merged_df.isnull().sum()
```

# Checking the unique values of all the necessary columns, their counts:

```python
# Specify columns for which you want to see unique values and counts
columns_to_inspect = ['Content Type','Category','Reaction Type','Sentiment']

# Iterate over specified columns and print unique values with their counts
for column in columns_to_inspect:
    print("Unique values and counts in column:", column)
    print(merged_df[column].value_counts())
    print()
```

*** Note:*** In the 'Category' column there are many inconsistent values and duplicates. I want to first remove the quotation marks(" ") and convert all the values to lower case to create consistency in this case.

```python
# Remove quotation marks and convert to lower case
merged_df['Category'] = merged_df['Category'].str.replace('"', '').str.lower()

merged_df['Category'].value_counts()
```

# Checking and correcting all the datatypes:

```python
merged_df.info()
```

# Fixing the datetime format:

```python
# Convert 'Datetime' column to datetime format
merged_df['Datetime'] = pd.to_datetime(merged_df['Datetime'])

merged_df['Datetime']
```

# Finally I'll Save the cleaned dataset into an Excel file format:

```python
# Define the file path where you want to save the Excel file
file_path = 'cleaned_data.xlsx'

# Save the DataFrame to an Excel file
merged_df.to_excel(file_path, index=False, header = True)

print("DataFrame saved to Excel successfully.")
```

# Checking the top reaction scores for all content categories:

```python
# Group by 'Category' and calculate the sum of Reaction Score for each category
category_reaction_sum = merged_df.groupby('Category')['Score'].sum()

# Sort the results in descending order
category_reaction_sum = category_reaction_sum.sort_values(ascending=False)

category_reaction_sum
```

# Saving it into a spreadsheet:

```python
# Convert the Series to a DataFrame
category_reaction_df = category_reaction_sum.reset_index(name='Sum of Reaction Scores')

# Define the file path where you want to save the Excel file
file_path = 'category_reaction_sum.xlsx'

# Save the DataFrame to an Excel file
category_reaction_df.to_excel(file_path, index=True)

print("Category reaction sum saved to Excel successfully.")
```

# Top Content Type:

```python
# Group by 'Category' and calculate the sum of Reaction Score for each category
content_reaction_sum = merged_df.groupby('Content Type')['Score'].sum()

# Sort the results in descending order
content_reaction_sum = content_reaction_sum.sort_values(ascending=False)

content_reaction_sum
```

```python
# Convert the Series to a DataFrame
content_reaction_df = content_reaction_sum.reset_index(name='Sum of Reaction Scores')

# Define the file path where you want to save the Excel file
file_path = 'content_reaction_sum.xlsx'

# Save the DataFrame to an Excel file
content_reaction_df.to_excel(file_path, index=True)

print("Content reaction sum saved to Excel successfully.")
```
